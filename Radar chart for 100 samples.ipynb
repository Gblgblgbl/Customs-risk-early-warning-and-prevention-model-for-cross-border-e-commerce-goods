```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 设置权重和标签
weights = np.array([0.06294601, 0.06190906, 0.06263205, 0.06233453, 0.06330322,
       0.06316696, 0.0629336 , 0.06258831, 0.06241387, 0.05405816,
       0.06167129, 0.00692863, 0.06316386, 0.06225412, 0.06267545,
       0.06248615, 0.06253473])
labels = [f'c{i+1}' for i in range(len(weights))]

# 检查权重和
if not np.isclose(weights.sum(), 1):
    raise ValueError("权重之和必须为1")

# 计算每个扇形的角度范围
angles = 2 * np.pi * np.cumsum(weights)  # 累加得到每个权重的终止角度
angles = np.concatenate(([0], angles))  # 起始角为0

# 计算扇形角平分线（即每个扇形区的中心角度）
angles_mid = (angles[:-1] + angles[1:]) / 2
angles_mid = np.concatenate((angles_mid, [angles_mid[0]]))
# 创建雷达图
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})

# 绘制参考圆
radius_65 = [0.65] * len(labels)
radius_85 = [0.85] * len(labels)
radius_65 += radius_65[:1]  # 闭合
radius_85 += radius_85[:1]  # 闭合

ax.plot(angles_mid, radius_65, linestyle='-', color='blue', label='0.65', linewidth=1)
ax.plot(angles_mid, radius_85, linestyle='-', color='red', label='0.85', linewidth=1)

# 设置刻度环
ax.set_ylim(0, 1)
ax.set_yticks([0.2, 0.4, 0.6, 0.8])
ax.set_yticklabels(['0.2', '0.4', '0.6', '0.8'], fontsize=10)
ax.grid(color='gray', linestyle='--', linewidth=0.5)

# 添加标签
ax.set_xticks(angles_mid[:-1])
ax.set_xticklabels(labels, fontsize=10)

# 添加标题
plt.title("Radar Chart (Sectors Centered on Axes)", va='bottom', fontsize=14)
plt.legend(loc='upper right')
plt.show()
```

```python
import numpy as np

# 假设 data2 是一个二维数组，形状为 (954, 17)，其中 954 是样本数，17 是特征数（C1-C17）
# data2 = np.array([...])  # 你的数据

# 权重
w = np.array([0.06294601, 0.06190906, 0.06263205, 0.06233453, 0.06330322,
       0.06316696, 0.0629336 , 0.06258831, 0.06241387, 0.05405816,
       0.06167129, 0.00692863, 0.06316386, 0.06225412, 0.06267545,
       0.06248615, 0.06253473])

# 初始化结果存储
results = []

# 逐行计算
for i in range(data2.shape[0]):
    row = data2[i]
    
    # 计算 S_i 和 L_i
    S_i = np.sum(row * np.sin(np.pi * w))
    L_i = np.sum(2 * np.sqrt((np.sin(np.pi/2 - np.pi * w) - row)**2 + np.sin(np.pi * w)**2))
    
    # 计算 S_m 和 L_m
    S_m = np.sum(np.sin(np.pi * w))
    L_m = np.sum(4 * np.sin(np.pi * w / 2))
    
    # 计算 V_i1, V_i2, 和 f_i
    V_i1 = S_i / S_m
    V_i2 = (S_i * L_m**2) / (L_i**2 * S_m)
    f_i = np.sqrt(V_i1 * V_i2)
    
    # 存储结果
    results.append(f_i)

# 转换为数组
results = np.array(results)

# 打印结果
print("Example1-100")
print("Results:", results)
```
